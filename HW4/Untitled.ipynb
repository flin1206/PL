```python
import openai
import pandas as pd
```

```python
keyfile = open("key.txt", "r")
key = keyfile.readline()
openai.api_key = key
```

```python
def chatgptfn(sub_list):
    result = ''
    print(sub_list)
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are an assistant"},
            {"role": "user", "content": f"{sub_list} :give me a summary"}
        ]
    )
    for choice in response.choices:
        result += choice.message.content
    return result
```

```python
with open('01.txt', 'r') as fh:
    tmp = fh.read()
    itemlist1 = tmp.split(',')
with open('02.txt', 'r') as fh:
    tmp = fh.read()
    itemlist2 = tmp.split(',')
```

```python
itemlist1 = str(itemlist1)
itemlist2 = str(itemlist2)
```

```python
data1 = [itemlist1[i:i+3200] for i in range(0, len(itemlist1), 3200)]
data2 = [itemlist2[i:i+3200] for i in range(0, len(itemlist2), 3200)]
```

```python
# 資料清洗
data1 = [chatgptfn(itemlist1[i:i+3200]) for i in range(0, len(itemlist1), 3200)]
data2 = [chatgptfn(itemlist2[i:i+3200]) for i in range(0, len(itemlist2), 3200)]
```

```python
data_1 = ''.join(data1)
data_2 = ''.join(data2)
papers = [data_1, data_2]
```

```python
data_1
```

```python
import gensim
from gensim import corpora
from pprint import pprint
```

```python
# 創建詞袋
texts = [[word for word in document.lower().split()] for document in papers]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
texts
```

```python
# 訓練 LDA 模型
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, 
                                            num_topics=2, random_state=100, update_every=1, 
                                            chunksize=100, passes=10, alpha='auto', per_word_topics=True)

# 輸出主題模型分析結果
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```
